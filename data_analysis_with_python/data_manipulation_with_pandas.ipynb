```python
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
print(df.head())
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().any()
```

```python
#handling missing values
df.isnull().sum()
```

```python
df_fillna = df.fillna(0)
print(df_fillna)
```

```python
#filling missing values with mean of the column
df['Sales_fillna_mean'] = df['Sales'].fillna(df['Sales'].mean())
print(df)
```

```python
#rename the columns
df = df.rename(columns={'Sales':'Total_Sales'})
print(df.head())
```

```python
#change data types
df['new_values'] = df['Value'].fillna(df['Value'].mean()).astype('int')
df.head()
```

```python
df['multiplied_values'] = df['Value'].apply(lambda x: x * 2)
df.head()
```

```python
##Data aggregation and grouping
grouped_mean = df.groupby('Product')['Total_Sales'].mean()
print(grouped_mean)
```

```python
df.groupby(['Product', 'Region'])['Value'].sum()
```

```python
df.groupby(['Product', 'Region'])['Value'].mean()
```

```python
#multiple aggregate fuctions
grouped_egg = df.groupby('Region')['Value'].agg(['sum', 'mean', 'max', 'count'])
print(grouped_egg)
```

```python
#Marge and Join dataFrames
df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie']
})
df2 = pd.DataFrame({
    'ID': [1, 2, 4],
    'Age': [25, 30, 35]
})
merged_df = pd.merge(df1, df2, on='ID', how='inner')
print(merged_df)
```

```python
merged_df = pd.merge(df1, df2, on='ID', how='outer')
print(merged_df)
```
