```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
pizza_size = np.arange(8, 13)
pizza_size
```

```python
pizza_price = np.arange(300, 550, 50)
pizza_price
```

```python
plt.scatter(pizza_size, pizza_price)
plt.title('Pizza Price Chart')
plt.xlabel('Size of Pizza (inches)')
plt.ylabel('Price of Pizza')
plt.show()
```

Line Chart


```python
plt.plot(pizza_size, pizza_price, marker='o', linestyle='--', color='b')
plt.xlabel('Size of Pizza (inches)')
plt.ylabel('Price of Pizza')
plt.title('Pizza Price Chart')
plt.show()
```

```python
blood_group = ['A+', 'B+', 'O+', 'AB+', 'A-', 'B-', 'O-', 'AB-']
person = [34, 23, 45, 12, 9, 5, 29, 7]
plt.bar(blood_group, person, color='purple', width=0.5)
plt.title('Blood Group Distribution')
plt.xlabel('Blood Group')
plt.ylabel('Number of Persons')
plt.show()
```

```python
plt.barh(blood_group, person, color='purple')
plt.title('Blood Group Distribution')
plt.xlabel('Blood Group')
plt.ylabel('Number of Persons')
plt.show()
```

```python
plt.pie(person, labels=blood_group, startangle=140, autopct = '%1.1f%%')
plt.title('Blood Group Distribution')
plt.show()
```

```python
x = np.random.randn(1000)
```

```python
a = np.random.randint(100, 500, 50)
a
```

```python
x = np.arange(0,10)
x
y = x**2
y
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,0.7,0.7])
axes.plot(x, y, color='red', marker='o', linestyle='-.')
axes.set_xlabel('X Axis Label')
axes.set_ylabel('Y Axis Label')
axes.set_title('Title of the Plot')
plt.show()
```

```python
fig.savefig('my_first_plot.png', bbox_inches='tight')
```

```python
month = ['January', 'February', 'March', 'April', 'May', 'June']
sales_2023 = [2500, 2700, 3000, 2800, 3200, 3500]
sales_2024 = [2600, 2900, 3100, 3000, 3300, 3600]
sales_2025 = [2700, 3000, 3200, 3100, 3400, 3700]
```

```python
plt.figure()
plt.plot(month, sales_2023, marker='o', label='2023 Sales')
plt.plot(month, sales_2024, marker='s', label='2024 Sales')
plt.plot(month, sales_2025, marker='^', label='2025 Sales')
plt.xlabel('Month')
plt.ylabel('Sales Amount ($)')
plt.title('Monthly Sales Comparison (2023-2025)')
plt.legend()
fig = plt.gcf()

plt.title('Monthly Sales Comparison (2023-2025)')
plt.show()
```

```python
fig.savefig('sales_comparison.png', bbox_inches='tight')
```
