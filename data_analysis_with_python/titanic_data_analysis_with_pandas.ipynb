```python
def findAge(age):
    if age < 18:
        return 1
    elif age >= 18 and age < 40:
        return 2
    elif age >= 40 and age < 60:
        return 3
    else:
        return 4
df['Age_group'] = df['Age'].apply(lambda x: findAge(x))
df.head()
```
