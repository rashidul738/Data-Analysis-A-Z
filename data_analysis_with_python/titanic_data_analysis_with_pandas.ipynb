```python
import pandas as pd
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df['Pclass'].value_counts()
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
df.head()
```

```python
df['first_name'] = df['Name'].apply(lambda x: x.split(', ')[1])
```

```python
df['last_name'] = df['Name'].apply(lambda x: x.split(', ')[0])
```

```python
df.head()
```

```python
df['first_name'].apply(lambda x: x.split('.')[1].strip())
```

```python
df.head()
```

```python
(df['Sex']==1)
```

```python
df['Third_&_man'] = ((df['Pclass']==3) & (df['Sex']==0)).astype(int)
```

```python
df.head()
```

```python
def findAge(age):
    if age < 18:
        return 1
    elif age >= 18 and age < 40:
        return 2
    elif age >= 40 and age < 60:
        return 3
    else:
        return 4
df['Age_group'] = df['Age'].apply(lambda x: findAge(x))
df.head()
```

```python
df = df.drop(['PassengerId'], axis=1)
df.head()
```

```python
df.rename(columns={'Name':'Full_name', 'Sex': 'Gender'}, inplace=True)
```
