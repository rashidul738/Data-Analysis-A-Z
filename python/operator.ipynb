###Arithmetic Operators (+, -, *, /, %, **, //)


```python
a = 10
b = 5
add_result = a + b
print("Addition:", add_result)
sub_result = a - b
print("Subtraction:", sub_result)
mul_result = a * b
print("Multiplication:", mul_result)
div_result = a / b
print("Division:", div_result)
mod_result = a % b
print("Modulus:", mod_result)
exp_result = a ** b
print("Exponentiation:", exp_result)
floor_div_result = a // b
print("Floor Division:", floor_div_result)
```

###Comparison Operators (==, !=, >, <, >=, <=)


```python
a=10
b=5
result = a == b
print("a == b:", result)
result = a != b
print("a != b:", result)
result = a > b
print("a > b:", result)
result = a < b
print("a < b:", result)
result = a >= b
print("a >= b:", result)    
result = a <= b
print("a <= b:", result)
```

###Logical Operators (and, or, not)

```python
a = 10
b = 5

# Logical AND
result = (a > 5) and (b < 10)
print("Logical AND two is correct:", result)
result = (a > 5) and (b > 10)
print("Logical AND one is correct:", result)
result = (a < 5) and (b > 10)
print("Logical AND two is incorrect:", result)

# Logical OR
result = (a > 5) or (b < 10)
print("Logical OR two is correct:", result)
result = (a > 5) or (b > 10)
print("Logical OR one is correct:", result)
result = (a < 5) or (b > 10)
print("Logical OR two is incorrect:", result)

# Logical NOT
result = not(a > 5)
print("Logical NOT:", result)
result = not(a < 5)
print("Logical NOT:", result)
```
