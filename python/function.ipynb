```python
def even_or_odd(num):
    if num%2 == 0:
        return "Even"
    else:
        return "Odd"
print(even_or_odd(6))  # Output: Odd
```

```python
#function with multiple parameters
def add_numbers(a, b):
    if a%2 == 0 and b%2 == 0:
        print("Both numbers are even")
        return a + b
    elif a%2 != 0 and b%2 == 0:
        print("First number is odd, second is even")
        return a - b
    elif a%2 == 0 and b%2 != 0:
        print("First number is even, second is odd")
        return b / a
    else:
        print("One or both numbers are odd")
        return a * b
print(add_numbers(3, 5))  # Output: 8
```

```python
def tempareture_conversion(tem, unit):
    if unit == "C":
        fahrenheit = (tem * 9/5) + 32
        return f"{tem}°C is {fahrenheit}°F"
    elif unit == "F":
        celsius = (tem - 32) * 5/9
        return f"{tem}°F is {int(celsius)}°C"
    else:
        return "Invalid unit. Please use 'C' for Celsius or 'F' for Fahrenheit."
        
```

```python
print(tempareture_conversion(72, 'F'))
```

```python
def is_strong_password(password):
    if len(password) < 8:
        return False
    has_upper = any(char.isupper() for char in password)
    has_lower = any(char.islower() for char in password)
    has_digit = any(char.isdigit() for char in password)
    has_special = any(not char.isalnum() for char in password)
    return has_upper and has_lower and has_digit and has_special
    return True

```

```python
is_strong = is_strong_password("Password123$")
print(is_strong)  # Output: True
```

```python
def calculatte_total_cost(cart):
    total = 0
    for item, price in cart.items():
        total += price
    return total
shopping_cart = {
    "apple": 1.2,
    "banana": 0.8,
    "orange": 1.5
}
total_cost = calculatte_total_cost(shopping_cart)
print(f"Total cost: ${total_cost:.2f}")
```

```python
def is_palindrome(s):
    s = s.replace(" ", "").lower()
    return s == s[::-1]
print(is_palindrome("A man a plan a canal Panama"))  # Output: True
print(is_palindrome("Hello"))  # Output: False
```

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
print(factorial(5))  # Output: 120
```

```python
def count_word_frequency(text):
    words = text.split()
    frequency = {}
    for word in words:
        word = word.lower().strip('.,!?;"\'()')
        if word in frequency:
            frequency[word] += 1
        else:
            frequency[word] = 1
    return frequency
text = "Hello world! Hello everyone. Welcome to the world of Python."
word_freq = count_word_frequency(text)
print(word_freq)
```
